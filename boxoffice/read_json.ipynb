```python
import pandas as pd
import json
import numpy as np

box_office = pd.read_json("box_office.json")

box_office
```

```python
box_office_exploded = box_office.explode(["ranks", "film_titles", "film_links", "worldwide_gross", "domestic_gross", "foreign_gross"])

box_office_exploded.columns = ["url", "rank", "title", "link", "worldwide_gross", "domestic_gross", "foreign_gross"]

for col in ["worldwide_gross", "domestic_gross", "foreign_gross"]:
    box_office_exploded[col] = box_office_exploded[col].apply(lambda x: x.replace("$", "").replace(",", "")).replace("-", np.nan).astype(float)

box_office_exploded

```
